```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=[1,2,563,4,5]
y=[10,15,13,18,16]
y1=[24,16,9,4,1]

plt.plot(x,y)
plt.plot(x,y1)
plt.title("Basic Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```
