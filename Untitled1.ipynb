```python
#fancy indexing
a = np.array([10,20,30,40,50])
indices = [0,2,4]
b=a[indices]
print("b",b)


#Boolean masking
mask=(a>30) & (a<50)
print("mask : ",mask)
c=a[mask]
print("c",c)
```

```python
# functions
# zerod
a=np.zeros((2,3),dtype="i")
print("a",a)
b=np.zeros((10),dtype="i")
print("b",b)

# ones
c=np.ones((2,3),dtype="i")
print("c",c)
d=np.ones((10),dtype="i")
print("d",d)

# arange
e=np.arange(1,10,2)
print("e",e)

# linespace

f=np.linspace(1,10,2)
print("f",f)
```
